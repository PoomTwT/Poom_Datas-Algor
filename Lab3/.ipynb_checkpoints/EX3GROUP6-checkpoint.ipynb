```python
import timeit
from timeit import Timer
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

#### len(data) - O(1)

```python
list_time=[]
tuple_time=[]
data=[]

for i in range(0,5001,2):
    testlist = list(range(i))
    testtuple = tuple(range(i))
    LT = Timer('len(testlist)','from __main__ import testlist' )
    TUP = Timer('len(testtuple)','from __main__ import testtuple' )
    data.append(i)
    list_time.append(LT.timeit(number=1000))
    tuple_time.append(TUP.timeit(number=1000))

table = list(zip(data, list_time , tuple_time))
print (pd.DataFrame(table , columns = ['Data Total', 'List', 'Tupple' ]) )

plt.plot(data,list_time,label='List')
plt.plot(data,tuple_time,label='Tupple')
plt.ylabel('times (ms)')
plt.xlabel('range of data')
plt.ylim(0,0.002)
plt.legend()
plt.show()
```

#### data1 + data2 - O(n1 + n2)

```python
#n1 = n2 ; O(2n)
list_time=[]
tuple_time=[]
empty_list_time = []
empty_tup_time = []

data=[]

t_list_e1 = [20]
t_list_e2 = [99]

t_tup_e1 = tuple([20])
t_tup_e2 = tuple([99])

for i in range(0,1001,2):
    testlist1 = list(range(i))
    testlist2 = list(range(i))
    
    testtuple1 = tuple(range(i))
    testtuple2 = tuple(range(i))

    data.append(i) 
    LT1 = Timer('testlist1 + testlist2','from __main__ import testlist1,testlist2' )
    TUP1 = Timer('testtuple1 + testtuple2','from __main__ import testtuple1,testtuple2' )
    
    LT2 = Timer('t_list_e1 + t_list_e2','from __main__ import t_list_e1,t_list_e2' )
    TUP2 = Timer('t_tup_e1 + t_tup_e1','from __main__ import t_tup_e1,t_tup_e2' )
    
    empty_list_time.append(LT2.timeit(number=1000))
    empty_tup_time.append(TUP2.timeit(number=1000))
    
    list_time.append(LT1.timeit(number=1000))
    tuple_time.append(TUP1.timeit(number=1000))

table1 = list(zip(data, list_time , tuple_time))
table2 = list(zip(data, empty_list_time , empty_tup_time))
print ("Table1 - n1=n2=n\n" ,pd.DataFrame(table1 , columns = ['Data Total', 'List', 'Tupple' ]))
print ("\nTable2 - n1,n2=0\n" ,pd.DataFrame(table2 , columns = ['Data Total', 'List', 'Tupple' ]))

plt.plot(data,list_time,label='List - (n1=n2)')
plt.plot(data,tuple_time,label='Tupple - (n1=n2)')

plt.plot(data,empty_list_time,label='List - (n1,n2 = 1)')
plt.plot(data,empty_tup_time,label='Tupple - (n1,n2 = 1)')

plt.ylabel('times (ms)')
plt.xlabel('range of data')
plt.legend()
plt.show()
```

#### c * data - O(cn)

```python
list_time=[]
tuple_time=[]

list_time_2=[]
tuple_time_2=[]
data=[]

for i in range(1,1000,10):
    testlist = list(range(i))
    testlist2 = list(range(i+1))
    testtuple = tuple(testlist)
    testtuple2 = tuple(testlist2)

    x = Timer(f'testlist*2','from __main__ import testlist,testlist2' )
    x_2 = Timer(f'testlist*3','from __main__ import testlist,testlist2' )
    y = Timer(f'testtuple*2','from __main__ import testtuple,testtuple2' )
    y_2 = Timer(f'testtuple*3','from __main__ import testtuple,testtuple2' )
    data.append(i)
    list_time.append(x.timeit(number=1000))
    list_time_2.append(x_2.timeit(number=1000))
    tuple_time.append(y.timeit(number=1000))
    tuple_time_2.append(y_2.timeit(number=1000))

display1 = list(zip(data, list_time,tuple_time))
display2 = list(zip(data, list_time_2,tuple_time_2))


df1 = pd.DataFrame(display1 , columns = ['Data', 'List*2','Tupple*2']) 
df2 = pd.DataFrame(display2 , columns = ['Data', 'List*3','Tupple*3'])

print(df1,'\n')
print(df2,'\n')

plt.plot(data,list_time,label=' 2*[list]')
plt.plot(data,list_time_2,label='3*[list]')
plt.plot(data,tuple_time,label='2*[Tuple]')
plt.plot(data,tuple_time_2,label='3*[Tuple]')

plt.ylabel('times')
plt.xlabel('range of n')
plt.legend()
plt.show()
```

#### data.append(value) - O(1)*

```python
list_time=[]
data=[]
testlist = []
size = []
for i in range(0,1001,1):
    x = Timer(lambda: testlist.append(i),'from __main__ import testlist' )
    data.append(i)
    list_time.append(x.timeit(number=1000))
    size.append(sys.getsizeof(testlist))

display1 = list(zip(data, list_time,size))
df1 = pd.DataFrame(display1 , columns = ['Data', 'Time','Size']) 
print(df1,'\n\n')

amt_list = []
amt_data = []
for i in range(len(size)):
    if size[i] != size[i-1]:
        amt_list.append(list_time[i])
        amt_data.append(data[i])
plt.plot(data,list_time,label='List')
plt.plot(amt_data,amt_list,label='Amortized')
plt.ylabel('times (ms)')
plt.xlabel('range of n')
plt.legend()
plt.show()
```
