```python
import timeit
from timeit import Timer
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
class LinkedStack:

        class _Node:

            __slots__ = '_element', '_next'

            def __init__(self, element, nextt):
                self._element = element
                self._next = nextt

        def __init__(self):
            self._head = None
            self._size = 0

        def __len__(self):
            return self._size

        def is_empty(self):
            return self._size == 0

        def push(self, e):
            self._head = self._Node(e, self._head)
            self._size += 1

        def top(self):
            if self.is_empty():
                raise Empty('Stack is empty')
            return self._head._element

        def pop(self):
            if self.is_empty():
                raise Empty('Stack is empty')
            answer = self._head._element
            self._head = self._head._next
            self._size -= 1
            return answer
```

```python
class LinkedQueue:
    
    class _Node:
        __slots__ = '_element', '_next'         # streamline memory usage
        
        def __init__(self, element, next):
            self._element = element
            self._next = next

    def __init__(self):
        self._head = None
        self._tail = None
        self._size = 0
    
    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0

    def first(self):
        if self.is_empty():
            raise Empty('Queue is empty')
        return self._head._element              # front aligned with head of list

    def dequeue(self):
        if self.is_empty():
            raise Empty('Queue is empty')
        answer = self._head._element
        self._head = self._head._next
        self._size -= 1
        if self.is_empty():
            self._tail = None
        return answer

    def enqueue(self, e):
        newest = self._Node(e, None)
        if self.is_empty():
            self._head = newest
        else:
            self._tail._next = newest
        self._tail = newest
        self._size += 1
```

# push and enqueue

```python
st2l = LinkedStack()
st2l_time =[]
q2l = LinkedQueue()
q2l_time =[]
data = []

for i in range(1,10000):
    x = Timer(lambda: q2l.enqueue(i),'from __main__ import q2l')
    y = Timer(lambda: st2l.push(i),'from __main__ import st2l')
    data.append(i)
    q2l_time.append(x.timeit(number=1000))
    st2l_time.append(y.timeit(number=1000))

plt.plot(data,q2l_time,label='queue')
plt.plot(data,st2l_time,label='stack')
plt.ylabel('Time Taken')
plt.xlabel('Data Size')
plt.ylim(0,0.03)
plt.legend()
plt.show()
```

```python
plt.plot(data,st2l_time,label='stack')
plt.ylabel('Time Taken')
plt.xlabel('Data Size')
plt.ylim(0,0.03)
plt.legend()
plt.show()
```

```python
plt.plot(data,q2l_time,label='queue')
plt.ylabel('Time Taken')
plt.xlabel('Data Size')
plt.ylim(0,0.03)
plt.legend()
plt.show()
```

```python
display1 = list(zip(data, st2l_time,q2l_time))
df1 = pd.DataFrame(display1, columns =  ['q2l', 'Stack','Queue'])
pd.set_option("display.max_rows",None,"display.max_columns",None)
print(df1)
```

# pop and dequeue

```python
q2l = LinkedQueue()
q2l_time =[]
st2l = LinkedStack()
st2l_time =[]
data = []

for i in range(1,1000000):
    q2l.enqueue(0)
    st2l.push(0)


for i in range(1,1000):
    x = Timer(lambda: q2l.dequeue(),'from __main__ import q2l')
    y = Timer(lambda: st2l.pop(),'from __main__ import st2l')
    data.append(i)
    q2l_time.append(x.timeit(number=1000))
    st2l_time.append(y.timeit(number=1000))
    
plt.plot(data,q2l_time,label='queue')
plt.plot(data,st2l_time,label='stack')
plt.ylabel('Time Taken')
plt.xlabel('Data Size')
plt.ylim(0,0.005)
plt.legend()
plt.show()
```

# top and first
