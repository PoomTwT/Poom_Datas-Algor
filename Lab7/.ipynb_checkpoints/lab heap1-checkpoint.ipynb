```python
a = [3,1,6,5,2,4,7]
def hps(a):
    for h in range(0,-len(a),-1): # 0 -1 -2 -3 -4 -5 -6
        for n in range(0,len(a)+h): # 0 1 2 3 4 5 6
            r = n
            L = 2*n + 1
            R = 2*n + 2
            if L < len(a) and a[L] > a[r]:
                r = L
                if h <= -1 and a[L] in a[h:]:
                    continue
                elif h <= -1 and a[R] in a[h:]:
                    #print(a[R],a[h:])
                    a[n],a[r] = a[r],a[n]
                    #print(a,'เปลี่ยนลูกซ้าย ลูกขวา sort แล้ว')
                    continue
            if R < len(a) and a[R] > a[r]:
                r = R
                if h <= -1 and a[r] in a[h:]:
                    #print(a,'ลูกขวา sort แล้ว')
                    continue
            a[n],a[r] = a[r],a[n]
            if a[n] > a[(n-1)//2]: #เช็คพ่อแม่
                a[n],a[(n-1)//2] = a[(n-1)//2],a[n]
            #print(a,'n=',n,'h=',h)

        a[0],a[len(a)+h-1] = a[len(a)+h-1],a[0]
        #print(a,'สลับเสด')
    return a
hps(a)
```
