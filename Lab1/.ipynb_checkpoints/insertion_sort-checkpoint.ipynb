```python
import timeit
from timeit import Timer
import matplotlib.pyplot as plt

def sort_s(func):
    name = func.__name__
    y = []
    x = []
    for i in range(1,3000,200):
        f1 = Timer(f"{name}([i for i in range({i})])", f"from __main__ import {name}")
        x.append(i)
        y.append(f1.timeit(number=1000))
    plt.plot(x, y, label= "sort_l") 
    plt.xlabel('range')
    plt.ylabel('time(s)')
    plt.legend()
    
import random
list_1 = [random.randint(-100,100) for i in range(1000)]
def sort_rd(func):
    name = func.__name__
    y = []
    x = []
    for i in range(1,1000,100):
        f1 = Timer(f"{name}({list_1[0:i]})", f"from __main__ import {name}")
        x.append(i)
        y.append(f1.timeit(number=1000))
    plt.plot(x, y, label= "random_l") 
    plt.xlabel('range')
    plt.ylabel('time(s)')
    plt.legend()
    
list_2 = [i for i in range(1000,0,-1)]
def sort_re(func):
    name = func.__name__
    y = []
    x = []
    for i in range(1,1000,100):
        f1 = Timer(f"{name}({list_2[0:i]})", f"from __main__ import {name}")
        x.append(i)
        y.append(f1.timeit(number=1000))
    plt.plot(x, y, label= "reverse_l") 
    plt.xlabel('range')
    plt.ylabel('time(s)')
    plt.legend()
    
    
def insertion_sort(array): #[0,1,2,3]
    
    for i in range(1, len(array)):
       
        key_item = array[i]
        j = i - 1
        
        while j >= 0 and array[j] > key_item:
            array[j + 1] = array[j]
            j -= 1

        array[j + 1] = key_item

    return array
```

```python
sort_s(insertion_sort)
#O(n)
```

```python
sort_rd(insertion_sort)
#O(n^2)
```

```python
sort_re(insertion_sort)
#O(n^2)
```

```python
sort_s(insertion_sort)
sort_rd(insertion_sort)
sort_re(insertion_sort)
```
