## hash function

```python
def hash_func1(w,m):
    mod = (m+1)*100
    ord_word = 0
    for string in w:
        ord_word += ord(string)
    return (ord_word%mod)
```

```python
def hash_func2(w,shift):
    word_shift = 0
    for string in w:
        word_shift = (word_shift << shift)
        word_shift += ord(string)
    return (word_shift)
#hash แบบ ง่าย สามารถเปลี่ยนได้ แต่แนะนำให้ใช้ชื่อ hash_func1 เหมือนเดิมจะได้ไม่ต้องไปแก้ตารางด้านล่าง
```

## ทดลอง

```python
import pandas as pd
```

```python
with open("words.txt", mode="r") as file :
    words = [i.strip() for i in file.readlines()]
```

```python
def find_collision(h_func,shift):
    col = list()
    for shifting in range(shift):
        theset = set()
        for w in words:
            shifted = h_func(w,shifting)
            theset.add(shifted)
        col.append(len(words) - len(theset))
    return(col)
#หาการชนกัน (ฟังก์ชันที่จะทดลอง,ต้องการชิฟไปเท่าไหร(เริ่มจาก0คือไม่ชิฟเลย))
```

```python
table = list(zip(range(50),find_collision(hash_func1,50) , find_collision(hash_func2,50)))
print (pd.DataFrame(table, columns = ['Shift','Collision-Hash1','Collision-Hash2']))
```

```python
#จากตัวอย่าง อธิบายจารได้ว่า ฟังชั่นแฮชแบบที่1 ควรปรับ m เป็น 23 ขึ้นไป จึงจะมีการชนกันน้อยที่สุดเท่าที่เป็นไปได้
#และ Hash 2 ควรปรับพารามิเตอร์shiftเป็น 7 ขึ้นไป เพราะทำให้ไม่เกิดการชนกัน
# แล้วเปรียบเทียบ Hash1 กับ Hash2 ได้ว่า Hash2 ดีกว่าเพราะ สามารถ ปรับพารามิเตอร์ shift แล้วทำให้ไม่เกิดการชนได้ แต่ Hash1 ปรับยังไงก็เกิดชนอยู่ดี
```
