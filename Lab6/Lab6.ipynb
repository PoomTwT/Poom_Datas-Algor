```python
def checkmid3(e):
    operator = ['+','-','*','^','%','/']
    l = []
    for i in range(len(e)):
        if e[i] in operator or  e[i]=='(' or  e[i] ==')':  
                #print(l,e[i])
            if( e[i] == '('):l.append(e[i])
            elif( e[i] == ')'):
                while(l[-1] != '('):l.pop()           
                l.pop()   
            else:
                l.append(i)
    return(l[0])

def countparan2(expr):
    stack = []
    l = []
    for i in expr:
        if i =='(' or i == ')':l.append(i)
    for i in range(len(l)-1):
        if l[i]==')' and l[i+1]=='(':
            l[i] = None
            l[i+1] = None
    for i in range(l.count(None)):
        l.remove(None)
    openpara = l.count('(')
    closepara = l.count(')')
    if (openpara+closepara)%2 == 0:
        return (openpara+closepara)
    else:
        raise ValueError
    
def Numberone(ex,floor,d):
   
    if(len(ex)==1):
        if floor in list(d.keys()):d[floor].append(ex)
        else:d[floor] = [ex]
        if floor+1 in list(d.keys()):
            d[floor+1] = d[floor+1]+[None,None]
        else:d[floor+1] =[None,None]
       # print(d)
        if len(d[floor]) != 2**(floor) and len(d[floor-1]) == 2**(floor-1):
            for i in range(len(d[floor-1])):
                if d[floor-1][i] == None:
                    d[floor].insert(2*i,None)
                    d[floor].insert(2*i+1,None)
        
        return (d) #ไม่ใช้ประโยชนแต่มีไว้เผื่อใช้       
    else:
        
        ex = ex[1:-1]
        mid = checkmid3(ex)
        root = ex[mid]
        L = ex[:mid] 
        R = ex[mid+1:]
        if floor in list(d.keys()):d[floor].append(root)
        else:d[floor] = [root] 
        left= Numberone(L,floor+1,d) #เรียกใช้ฟังชั่นแต่มีตัวมารับด้วย
        right= Numberone(R,floor+1,d)
        #print('left :',left, 'right:',right)
        
        return(d) 
#ans = Numberone('((4*5)+((1+2)+(1+1)))',0,{})
def buildTree(X):
    ans = Numberone(X,0,{})
    
    answer =[]
    for j in list(ans.keys()):
        if ans[j] == [None]*len(ans[j]):
            ans.pop(j)
    #print(ans)
    for i in list(ans.keys()):
        for j in ans[i]:
            if j == None:
                answer.append(None)
            else:answer += j
    tree=[]
    nfloor = countparan2(X)//2
    for i in range(nfloor+1):
        Ln=[None]*(2**i)
        tree = tree + Ln
    #print(tree)
    for i in range(len(answer)):
        tree[i] = answer[i]
    return tree


print(buildTree('(((1+4)*(5*7))+3)') )
print(buildTree('(3+((1+4)*(5*7)))') )
print(buildTree('((4*(5*7))+3)') )
print(buildTree('(3+(4*(5*7)))') )
print(buildTree('(3+(4*5))') )
print(buildTree('((4*5)+3)') )
print(buildTree('((1/3)+((1+4)*(5*7)))') )
print(buildTree('(9+(3*(9/3)))') )
```

```python
def reverseTree(ex):
    operator = ['+','-','*','^','%','/']
    floor = 0
    d={}
    for i in ex:
        if floor in list(d.keys()):d[floor].append(i)
        else:d[floor] = [i]
        if len(d[floor])== 2**floor:
            floor+=1
    for i in list(d.keys()):
        if d[i].count(None)>0:
            for j in range(len(d[i])):
                if d[i][j] == None: d[i][j]='0'
    return(d)
def listToString(s):   
    str1 = ""    
    for ele in s:  
        str1 += ele      
    return str1  

def calTree(ex):
    rt = reverseTree(ex) 
    key = list(rt.keys())
    for i in key[::-1]: #ย้อนหลังมาหน้า
        if i !=0:
            for k in range(0,len(rt[i]),2):
                operands = rt[i][k:k+2]
                operator = rt[i-1][k//2]
                exp =[operands[0]]+[operator]+[operands[-1]] #จัดรูป
                if not(exp[0] =='0' and exp[-1] =='0' and not(exp[1] in ['+','-','*','^','%','/'] )):
                    e = listToString(exp)
                    e = eval(e)
                    rt[i-1][k//2] = str(e)
    return(e)  


print(calTree(buildTree('(((1+4)*(5*7))+3)')))
print(calTree(buildTree('(3+((1+4)*(5*7)))')))
print(calTree(buildTree('((4*(5*7))+3)')))
print(calTree(buildTree('(3+(4*(5*7)))')))
print(calTree(buildTree('(3+(4*5))')))
print(calTree(buildTree('((4*5)+3)')))
print(calTree(buildTree('((1/3)+((1+4)*(5*7)))')))
print(calTree(buildTree('(9+(3*(9/3)))')))
```

```python
class BinaryTree:
    
    def __init__(self, root):
        self.key = root
        self.left_child = None
        self.right_child = None
        
    def insert_left(self, new_node):
        if self.left_child == None:
            self.left_child = BinaryTree(new_node)
        else:
            t = BinaryTree(new_node)
            t.left_child = self.left_child
            self.left_child = t
            
    def insert_right(self, new_node):
        if self.right_child == None:
            self.right_child = BinaryTree(new_node)
        else:
            t = BinaryTree(new_node)
            t.right_child = self.right_child
            self.right_child = t
        
    def get_right_child(self):
        return self.right_child
    
    def get_left_child(self):
        return self.left_child
    
    def set_root_val(self, obj):
        self.key = obj
        
    def get_root_val(self):
        return self.key
```

```python
Tree = BinaryTree('*')
Tree.insert_left('+')
Tree.insert_right('-')
Tree.get_left_child().insert_left('7')
Tree.get_left_child().insert_right('3')
Tree.get_right_child().insert_left('5')
Tree.get_right_child().insert_right('2')
```

```python
def tree_calculator(t):
    root = t.get_root_val()
    L = t.get_left_child()
    R = t.get_right_child()
    if L == None and R == None:
        return (int(root))
    else:
        if root == '+':
            return(tree_calculator(L) + tree_calculator(R))
        elif root == '-':
            return(tree_calculator(L) - tree_calculator(R))
        elif root == '*':
            return(tree_calculator(L) * tree_calculator(R))
        elif root == '/':
            return(tree_calculator(L) / tree_calculator(R))
        elif root == '^':
            return(tree_calculator(L) ** tree_calculator(R))
        elif root == '%':
            return(tree_calculator(L) % tree_calculator(R))
        else:
            raise ValueError
```
