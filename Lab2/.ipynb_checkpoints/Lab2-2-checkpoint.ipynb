## Harmonic Number

```python
def harmonic_number(n):
    if type(n) == int and n>=1 :

        if n == 1 :
            return 1
        elif not (n<1):
            return (1/n) + harmonic_number(n-1)

    else:
        return False
```

```python
harmonic_number(-10)
```

## Max - Min

```python
Set = [2,6,7,99,2,5,4,10,69,1,7,48]
```

```python
def find_m(S):
    if len(S) == 1:
        return (S[0],S[0])

    finding = find_m(S[1:]) #recursion 

    if S[0] >= finding[0]:
        Max = S[0]
    else:
        Max = finding[0]

    if S[0] <= finding[1]:
        Min = S[0]
    else:
        Min = finding[1]
    return (Max,Min)
```

```python
find_m(Set)
```

## Palindrome

```python
odd_pld_str = "abcdcba"
even_pld_str = "zyxxyz"
long_pld_str = "aaabbgggkkfkkUkkfkkgggbbaaa"
####
odd_unpld_str = "aacca"
even_unpld_str ="jkjj"
long_unpld_str = "GGJGaytDrHUrDtyaGJGG"
```

```python
def palindrome(string):
    if len(string) < 2:
        return True
    elif string[0] != string[-1]:
        return False
    else:
        return palindrome(string[1:-1]) #recursion
```

```python
print(palindrome(odd_pld_str))
print(palindrome(even_pld_str))
print(palindrome(long_pld_str),"\n************************")
####
print(palindrome(odd_unpld_str))
print(palindrome(even_unpld_str))
print(palindrome(long_unpld_str))
```
